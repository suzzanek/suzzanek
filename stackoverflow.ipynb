```python
def categorize_branch(x):
    if x == 'I am a developer by profession':
        return 'Developer'
    elif x == 'I am not primarily a developer, but I write code sometimes as part of my work/studies':
        return 'Work and Studies'
    elif x == 'I code primarily as a hobby':
        return 'Hobby'
    elif x == 'I am learning to code':
        return 'Learning'
    elif x == 'I used to be a developer by profession, but no longer am':
        return 'Not Working Anymore'
    else:
        return x 
    
df['Branch']= df['Branch'].apply(categorize_branch)
```

```python
def certifications(x):
    if pd.notnull(x):
        return "Yes"
    else:
        return x 
    
df['OnlineCertifications'] = df['OnlineCertifications'].apply(certifications)
```

```python
def age_category(x):
    if x == '18-24 years old':
        return 1
    elif x == 'Under 18 years old':
        return 0
    elif x == '25-34 years old':
        return 2
    elif x ==  '45-54 years old':
        return 3
    elif x == '35-44 years old':
        return 4
    
    elif x =='55-64 years old':
        return 5
    elif x =='65 years or older':
        return 6
    elif x =='Prefer not to say':
        return 'NaN'
    else:
        return x 
    
df['Age']= df['Age'].apply(age_category)
```
